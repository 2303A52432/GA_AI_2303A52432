```python
import numpy as np

# Training data (input x1, x2, x3 and output y)
train_data = np.array([
    [0.1, 0.2, 0.3, 0.14],
    [0.2, 0.3, 0.4, 0.20],
    [0.3, 0.4, 0.5, 0.26],
    [0.5, 0.6, 0.7, 0.38],
    [0.1, 0.3, 0.5, 0.22],
    [0.2, 0.4, 0.6, 0.28],
    [0.3, 0.5, 0.7, 0.34],
    [0.4, 0.6, 0.8, 0.40],
    [0.5, 0.7, 0.1, 0.22]
])

# Test data (input x1, x2, x3 and expected output y)
test_data = np.array([
    [0.6, 0.7, 0.8, 0.44],
    [0.7, 0.8, 0.9, 0.50]
])

# Initialize the weights and bias (random values)
weights = np.random.randn(3)  # 3 input features (x1, x2, x3)
bias = np.random.randn()  # 1 bias term

# Hyperparameters
learning_rate = 0.1  # Learning rate for gradient descent
epochs = 1000  # Number of iterations for training

# Linear activation function (output layer)
def linear_activation(x):
    return x

# Mean Squared Error (MSE) cost function
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Forward pass (predict the output)
def predict(X):
    return linear_activation(np.dot(X, weights) + bias)

# Backpropagation algorithm (gradient descent to update weights and bias)
def train_model(train_data, learning_rate, epochs):
    global weights, bias
    for epoch in range(epochs):
        # Extract inputs (X) and outputs (y)
        X_train = train_data[:, :-1]  # All columns except last (inputs)
        y_train = train_data[:, -1]  # Last column (outputs)

        # Perform forward pass
        y_pred = predict(X_train)

        # Calculate the error (derivative of the MSE cost function)
        error = y_pred - y_train

        # Compute gradients (derivatives with respect to weights and bias)
        grad_weights = (2 / len(X_train)) * np.dot(X_train.T, error)
        grad_bias = (2 / len(X_train)) * np.sum(error)

        # Update weights and bias using gradient descent
        weights -= learning_rate * grad_weights
        bias -= learning_rate * grad_bias

        # Optionally print the MSE every 100 epochs to see progress
        if epoch % 100 == 0:
            mse = mean_squared_error(y_train, y_pred)
            print(f"Epoch {epoch}, MSE: {mse}")

# Function to calculate MSE on the test dataset
def test_model(test_data):
    X_test = test_data[:, :-1]
    y_test = test_data[:, -1]
    y_pred = predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse

# Train the model
train_model(train_data, learning_rate, epochs)

# Test the model
mse_test = test_model(test_data)
print(f"Mean Squared Error on test data: {mse_test}")

# Function to predict output for user input
def predict_user_input():
    x1 = float(input("Enter value for x1: "))
    x2 = float(input("Enter value for x2: "))
    x3 = float(input("Enter value for x3: "))
    user_input = np.array([x1, x2, x3])
    prediction = predict(user_input)
    print(f"Predicted output: {prediction}")

# Make predictions based on user input
predict_user_input()
```

```python
import numpy as np

# Training data (x1, x2, x3, and y)
train_data = np.array([
    [0.1, 0.2, 0.3, 0.5349],
    [0.2, 0.3, 0.4, 0.5498],
    [0.3, 0.4, 0.5, 0.5646],
    [0.5, 0.6, 0.7, 0.5939],
    [0.1, 0.3, 0.5, 0.5548],
    [0.2, 0.4, 0.6, 0.5695],
    [0.3, 0.5, 0.7, 0.5842],
    [0.4, 0.6, 0.8, 0.5987],
    [0.5, 0.7, 0.1, 0.5548]
])

# Test data (x1, x2, x3, and y)
test_data = np.array([
    [0.6, 0.7, 0.8, 0.6083],
    [0.7, 0.8, 0.9, 0.6225]
])

# Initialize weights and bias randomly
weights = np.random.randn(3)  # Weights for x1, x2, x3
bias = np.random.randn()  # Bias term

# Hyperparameters
learning_rate = 0.1  # Learning rate for gradient descent
epochs = 1000  # Number of iterations for training

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Mean Squared Error (MSE) cost function
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Forward pass: Predict output
def predict(X):
    return sigmoid(np.dot(X, weights) + bias)

# Backpropagation algorithm: Gradient descent to update weights and bias
def train_model(train_data, learning_rate, epochs):
    global weights, bias
    for epoch in range(epochs):
        # Extract inputs (X) and outputs (y)
        X_train = train_data[:, :-1]  # All columns except last (inputs)
        y_train = train_data[:, -1]  # Last column (outputs)

        # Perform forward pass
        y_pred = predict(X_train)

        # Calculate the error (derivative of the MSE cost function)
        error = y_pred - y_train

        # Compute gradients (derivatives with respect to weights and bias)
        grad_weights = (2 / len(X_train)) * np.dot(X_train.T, error * sigmoid_derivative(y_pred))
        grad_bias = (2 / len(X_train)) * np.sum(error * sigmoid_derivative(y_pred))

        # Update weights and bias using gradient descent
        weights -= learning_rate * grad_weights
        bias -= learning_rate * grad_bias

        # Optionally print the MSE every 100 epochs to see progress
        if epoch % 100 == 0:
            mse = mean_squared_error(y_train, y_pred)
            print(f"Epoch {epoch}, MSE: {mse}")

# Function to calculate MSE on the test dataset
def test_model(test_data):
    X_test = test_data[:, :-1]
    y_test = test_data[:, -1]
    y_pred = predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    return mse

# Train the model
train_model(train_data, learning_rate, epochs)

# Test the model
mse_test = test_model(test_data)
print(f"Mean Squared Error on test data: {mse_test}")

# Function to predict output for user input
def predict_user_input():
    x1 = float(input("Enter value for x1: "))
    x2 = float(input("Enter value for x2: "))
    x3 = float(input("Enter value for x3: "))
    user_input = np.array([x1, x2, x3])
    prediction = predict(user_input)
    print(f"Predicted output: {prediction}")

# Make predictions based on user input
predict_user_input()
```
